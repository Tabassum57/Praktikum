```python
import os
from ultralytics import YOLO

# Verify if the file exists
model_path = r'C:\Users\hp\Desktop\limu\best.pt'  # Make sure this is the correct path to your model
video_path = r'C:\Users\hp\Desktop\limu\road.mp4'  # Path to your video file

# List files in the directory for debugging
print("Files in the directory:")
for file in os.listdir(r'C:\Users\hp\Desktop\lane_data'):
    print(file)

# Check if model file exists
if not os.path.exists(model_path):
    raise FileNotFoundError(f"Model file not found: {model_path}")

# Check if video file exists
if not os.path.exists(video_path):
    raise FileNotFoundError(f"Video file not found: {video_path}")

# Load the model
model = YOLO(model_path)

# Run inference on the source
model.predict(video_path, save=True, conf=0.5)
```
